# News Article Summarization using NLP

## Objective

This project automatically summarizes lengthy news articles into concise summaries using Natural Language Processing (NLP).

The system extracts the most important sentences from news articles and generates a shorter version while preserving key information.

## Technologies Used

- Python
- Pandas
- NLTK
- Scikit-Learn

```python
# Import Required Libraries

import pandas as pd
import nltk
import heapq
import re

from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
# Download NLTK Resources

nltk.download('punkt')
nltk.download('stopwords')
```

## Load Dataset

```python
# Read Dataset

df = pd.read_csv(
    "Articles.csv",
    encoding="latin1"
)

df.head()
```

## Dataset Information

```python
print("Dataset Shape :", df.shape)

print(df.columns)
```

```python
df.isnull().sum()
```

## Select News Article

```python
article = df["Article"][0]

print(article[:1000])
```

## Text Preprocessing

```python
article_text = re.sub(r'\[[0-9]*\]', ' ', article)

article_text = re.sub(r'\s+', ' ', article_text)

formatted_text = re.sub('[^a-zA-Z]', ' ', article_text)

formatted_text = re.sub(r'\s+', ' ', formatted_text)
```

## Calculate Word Frequencies

```python
stop_words = set(stopwords.words("english"))

word_frequencies = {}

for word in word_tokenize(formatted_text):

    if word.lower() not in stop_words:

        if word not in word_frequencies:

            word_frequencies[word] = 1

        else:

            word_frequencies[word] += 1
```

```python
maximum_frequency = max(word_frequencies.values())

for word in word_frequencies.keys():

    word_frequencies[word] = (
        word_frequencies[word] /
        maximum_frequency
    )
```

## Score Sentences

```python
sentence_scores = {}

sentences = sent_tokenize(article)

for sent in sentences:

    for word in word_tokenize(sent.lower()):

        if word in word_frequencies.keys():

            if len(sent.split(' ')) < 35:

                if sent not in sentence_scores:

                    sentence_scores[sent] = word_frequencies[word]

                else:

                    sentence_scores[sent] += word_frequencies[word]
```

## Generate Summary

```python
summary_sentences = heapq.nlargest(
    5,
    sentence_scores,
    key=sentence_scores.get
)

summary = ' '.join(summary_sentences)

print(summary)
```

## Compare Length

```python
print(
    "Original Length :",
    len(article)
)

print(
    "Summary Length :",
    len(summary)
)
```
