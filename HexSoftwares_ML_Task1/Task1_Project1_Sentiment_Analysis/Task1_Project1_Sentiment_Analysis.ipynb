# Sentiment Analysis using Machine Learning

## Project Objective

Sentiment Analysis is a Natural Language Processing (NLP) technique used to determine the emotional tone behind text.

This project classifies customer reviews into:

- Positive
- Negative
- Neutral

## Technologies Used

- Python
- Pandas
- Scikit-Learn
- Natural Language Processing (NLP)

## Dataset

The dataset contains customer reviews and their corresponding sentiments.

```python
# Import Required Libraries

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

## Load Dataset

```python
df = pd.read_csv("reviews.csv")
df.head()
```

```python
print("Dataset Shape :", df.shape)
print("\nColumns :")
print(df.columns)
```

## Dataset Information

```python
df.info()
```

```python
df.isnull().sum()
```

## Feature Selection

```python
X = df["Review"]
y = df["Sentiment"]
```

## Convert Text Data into Numerical Format

```python
vectorizer = CountVectorizer()
X_vectorized = vectorizer.fit_transform(X)
print(X_vectorized.shape)
```

## Split Dataset into Training and Testing Sets

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized,
    y,
    test_size=0.2,
    random_state=42
)

print("Training Data Shape :", X_train.shape)
print("Testing Data Shape :", X_test.shape)
```

## Train Machine Learning Model

```python
model = MultinomialNB()
model.fit(X_train, y_train)
print("Model Trained Successfully")
```

## Make Predictions

```python
y_pred = model.predict(X_test)

print(y_pred)
```

## Evaluate Model Performance

```python
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy Score :", accuracy)
```

```python
print(classification_report(y_test, y_pred))
```

## Test Custom Review

```python
review = ["This product is fantastic"]
review_vector = vectorizer.transform(review)
prediction = model.predict(review_vector)

print("Review :", review[0])
print("Predicted Sentiment :", prediction[0])
```

# Conclusion

The Sentiment Analysis model successfully predicts customer sentiment using Machine Learning and NLP techniques.

Algorithm Used:
- CountVectorizer
- Multinomial Naive Bayes

The model can classify reviews into:
- Positive
