# Music Recommendation System

## Objective

A Music Recommendation System suggests songs based on similarities in music data.

The system recommends songs according to:

- Genre
- User Preferences
- Similar Songs

## Technologies Used

- Python
- Pandas
- Scikit-Learn

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

## Load Dataset

```python
df = pd.read_csv("songs.csv")
df.head()
```

```python
print("Dataset Shape :", df.shape)
df.info()
```

## Dataset Preview

```python
df
```

## Convert Genre into Numerical Features

```python
vectorizer = CountVectorizer()

genre_matrix = vectorizer.fit_transform(df["Genre"])
```

## Calculate Similarity Matrix

```python
similarity = cosine_similarity(genre_matrix)

similarity
```

## Recommendation Function

```python
def recommend_song(song_name):

    song_index = df[df["Song"] == song_name].index[0]

    scores = list(enumerate(similarity[song_index]))

    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print("Recommended Songs:\n")

    for item in scores[1:6]:
        print(df.iloc[item[0]]["Song"])
```

## Test Recommendation System

```python
recommend_song("Believer")
```

```python
recommend_song("Tum Hi Ho")
```

# Conclusion

The Music Recommendation System successfully recommends songs based on genre similarity.
